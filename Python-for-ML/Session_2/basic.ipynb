```python
while True:
    degree = input("Enter your degree: ")
    if int(degree) <=100 and int(degree) >=0:
        degree = int(degree)
        break

if degree > 90:
    print("A")
elif degree > 80:
    print("B")
elif degree > 70:
    print("C")
elif degree > 60:
    print("D")    
else:
    print("F")
```
