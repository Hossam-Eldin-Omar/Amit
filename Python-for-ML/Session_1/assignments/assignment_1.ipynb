# Task-1 
#### Suppose you have this email address “Amit_ml@gmail.edu”
- Input Validation: Check if the input string contains exactly one "@" symbol and 
at least one "." after the "@" symbol. If it's not a valid email, return "Invalid 
email". 
- Extract Username: Extract and return the part of the email before the "@" 
symbol. 
- Extract Domain: Extract and return the domain (the part between "@" and the 
last ".").
- Check for Domain Ending: Check if the email ends with ".com". If it does, 
return "Commercial Domain". If it ends with ".edu", return "Educational Domain". 
Otherwise, return "Other Domain".

```python
email = "Amit_ml@gmail.edu"
if(email.count("@")==1):
    user, domain = email.split("@")
    if(domain.count(".")>=1):
        print("Valid Email")
        username = user
        print("Username:", username)
        domain_name, extension = domain.split(".", 1)
        print("Domain Name:", domain_name)
        if email.endswith(".com"):
            print("Commercial Domain")
        elif "edu" in extension:
            print("Educational Domain")
        else:
            print("Other Domain")
    else:
        print("Invalid Email: Domain must contain at least one '.'")
else:
    print("Invalid Email: Domain must contain exactly one '@'")
```

# Task-2:
- Encoded Message: ###!!@mocleW EPGTQ!!!6789
- Steps to Decode:
 
    1. Extract the core part of the message: "mocleW EPGTQ". 
    2. Reverse the first word: "mocleW" becomes "Welcome". 
    3. Replace shifted vowels in the second word: 
        - "EPGTQ": No vowels to change. 
    4. Final decoded message: "Welcome PGTQ". 

```python
encoded_msg1 = "###!!@mocleW EPGTQ!!!6789"
core = encoded_msg1.strip("@#!6789")
print(f"core part of the message: \"{core}\"")
first_half, second_half = core.split(" ")
first_half = first_half[::-1]
print(f"first half after reversing: \"{first_half}\"")
second_half = second_half[0].lower() + " " + second_half[1:]
print(f"Final decoded message: \"{first_half + second_half}\"")
```

# Task-3: 
- Encoded Message: &&&**$gnirtS PLIO!!@1234
- Steps to Decode:
 
    1. Extract the core part of the message: "gnirtS PLIO".
    2. Reverse the first word: "gnirtS" becomes "String".
    3. Replace shifted vowels in the second word:
        - "PLIO": Replace I->E and O->U to get "PLEU".
    4. Final decoded message: "String PLEU".

```python
encoded_msg2 = "&&&**$gnirtS PLIO!!@1234"
core = encoded_msg2.strip("&*$!@!1234")
print(f"core part of the message: \"{core}\"")
first_word, second_word = core.split(" ")
first_word = first_word[::-1]
print(f"first word after reversing: \"{first_word}\"")
second_word = second_word.replace("I", "E").replace("O", "U")
print(f"second word after vowerl shifting: \"{second_word   }\"")
print(f"Final decoded message: \"{first_word + " " + second_word}\"")
```
