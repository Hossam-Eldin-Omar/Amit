```python
import pandas as pd

# Load the dataset
df = pd.read_csv('train.csv')

# Create the chk_nulls function
def chk_nulls(df):
    """
    Analyze null values in a DataFrame.
    Returns a DataFrame with null counts and ratios for each column.
    """
    null_counts = df.isnull().sum()
    total_rows = len(df)
    null_ratios = df.isnull().sum() / total_rows
    
    result = pd.DataFrame({
        'Null': null_counts,
        'ratio': null_ratios
    })
    
    return result.T

# Run the analysis
result = chk_nulls(df)

# Display with all columns visible
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', None)

print(result)
```
