```python
def mult(x,y):
    """
    param_1: take first number as input
    type param_1: x: int
    param_2: take second number as input    
    type param_2: y: int
    ---
    returns: product of two numbers
    rtype: int
    """
    print("Mult function called with:", x*y)

def divide(x,y):
    """
    param_1: take first number as input
    type param_1: x: int
    param_2: take second number as input    
    type param_2: y: int
    ---
    returns: division of two numbers
    rtype: float
    """
    print("Divide function called with:", x/y)

def subtract(x,y):
    """
    param_1: take first number as input
    type param_1: x: int
    param_2: take second number as input    
    type param_2: y: int
    ---
    returns: difference of two numbers
    rtype: int
    """
    print("Subtract function called with:", x-y)

def add(x,y):
    """
    param_1: take first number as input
    type param_1: x: int
    param_2: take second number as input    
    type param_2: y: int
    ---
    returns: sum of two numbers
    rtype: int
    """
    print("Add function called with:", x+y)

if __name__ == "__main__":
    mult(10,20)
    divide(10,20)
    subtract(10,20)
    add(10,20)
```

```python
def fact(num):
    if num == 1 or num == 0:
        return 1
    return num * fact(num - 1)
fact(0)
```

```python
def isPrime(num):
    """
    """
    for i in range (2, num):
        if num%i == 0:
            return False
    return True
isPrime(17)
```

```python
def dividors(num1, num2):
    """
    param_1: take first number as input
    ---
    type param_1: x: int
    ---
    param_2: take second number as input    
    ---
    type param_2: y: int
    ---
    returns: the common dividors of the two numbers
    ---
    rtype: list
    """
    mn = min(num1,num2)
    cmn_div = []
    for i in range (1, mn+1):
        if num1%i == 0 and num2 %i == 0:
            cmn_div.append(i)
    return cmn_div

dividors(5,10)
```

```python
def even_fun(num):
    even_list = []
    for i in range (num):
        if i%2 == 0:
            even_list.append(i)
    return even_list


even_fun(10)
```
